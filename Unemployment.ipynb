```python
df2=pd.read_csv('Unemployment_Rate_upto_11_2020.csv')
df2
```

```python
df3= df2[['Region','Region.1']]
```

```python
df3['duplicated']=df3.duplicated()
df3 = df3[df3['duplicated'] == False]
df3.drop(columns='duplicated',inplace=True)
```

```python
df3
```

```python
df1 = pd.merge(df, df3[['Region', 'Region.1']], on=['Region'], how='left')
df1.shape
```

```python
df1.rename(columns={'Region': 'State','Region.1':'Region',df.columns[5] :'ELPR',df.columns[3]:'EUR',df.columns[4]:'EE'},inplace=True)
```

```python
df1.head()
```

```python
plt.pie(df1.groupby(by='Area')['EUR'].mean(),labels=df['Area'].unique(),autopct='%0.2f%%')
plt.tight_layout()
plt.show()
```

```python
sns.lineplot(df1.groupby('State')['EUR'].mean().sort_values(ascending= False))
plt.ylabel('Estimated Unemployment Rate')
plt.xticks(rotation=90)
plt.show()
```

```python
sns.lineplot(df1.groupby('State')['ELPR'].mean().sort_values(ascending= False))
plt.xticks(rotation=90)
plt.show()
```

```python
sns.barplot(df1.groupby('Region')['EUR'].mean().sort_values(ascending= False),color='purple')
plt.xticks(rotation=90)
plt.show()
```
